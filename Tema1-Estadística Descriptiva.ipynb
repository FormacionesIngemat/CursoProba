## Primeros pasos en R

R puede ser usado para realizar cálculos sencillos, al igual que una calculadora simple. Por ejemplo, si realizamos una operación aritmética básica, nos devuelve el resultado:

````r
2+2
````

Ahora bien, si queremos almacenar el resultado de una operación para usarlo después, podemos *asignar* el valor a una *variable*, de tal forma que podamos usar dicha variable en el futuro. Por ejemplo, calcularemos el resultado de multiplicar 2 por 3 y lo guardaremos en una variable a la que denominaremos ```A```. Esto se realiza con el operador de asignación ```<-```, a cuya izquierda va el nombre de la variable y a la derecha el valor a ser almacenado, como mostramos a continuación:



````r
A<-2*3
````

Así, si más adelante deseo conocer el valor de ```A```, solo debo escribir su nombre:

````r
A
````

O también puedo operar con él:

````r
A+5
````



> **Atención:** R distingue entre mayúsculas y minúsculas. De esta forma, si llamara a la variable ```a```, obtendría un error, puesto que no le he asignado ningún valor. Miremos a continuación lo que sucede en este caso.

````r
a
````

## Tipos de datos y objetos en R

Los tipos de datos más comunes con los que se trabaja en R son:

- *Entero* ```(integer)```: valor numérico que como su nombre indica, solo puede tomar valores enteros. Es útil para almacenar edades, contar objetos, etc.
- *Numérico* ```(numeric)```: este tipo de dato almacena números con decimales, y es el más usado en las operaciones cotidianas en R.
- *Texto* ```(character)```: permite almacenar palabras u oraciones completas, y manipularlas contando su longitud, reemplazando caracteres, etc. Es importante usar comillas al usar este tipo de dato, puesto que no es lo mismo ingresar ```25``` que ```"25"``` (el primero se entiende como unnúmero, el segundo como texto).
- *Lógico* ```(logical)```: solo toma dos valores: ```TRUE``` o ```FALSE```. Como su nombre indica, permite modelar características binarias de presencia o ausencia.

Existen otros tipos de datos más avanzados para propósitos precisos (fechas, factores...) pero para iniciar consideraremos los previamente definidos.

Además de estos tipos de datos, existen tipos de estructuras en las cuales almacenar los datos. Además de las variables, las cuales ya manejamos previamente, hay dos estructuras básicas que usaremos en este curso: los *vectores* y los *data frames.* Hablaremos ahora de los primeros.

### Vectores

Los *vectores* son arreglos unidimensionales de datos de un solo tipo. Se pueden declarar usando la función ```c()``` y separando con comas los elementos del vector. Por ejemplo, si quiero crear y almacenar un vector numérico llamado ```x``` con los números -5, 8, 3, 5, hago lo siguiente:

````r
x<-c(-5, 8, 3, 5)
x
````

O si quiero crear un vector con cadenas de texto:

````r
y<-c("a","bb","ccc","dddd")
y
````

Un fenómeno importante al trabajar con vectores en R es el de la **coerción**. Como mencionamos antes, los vectores solo permiten el uso de un tipo de datos. Por tanto, al ingresar un vector con varios tipos de datos, R transforma algunos para garantizar la uniformidad del vector. Por ejemplo, si introducimos un vector con diferentes tipos de datos

````r
z<-c(3,-3.2,"cinco",FALSE)
z
````

notamos que ha transformado todos los elementos a texto. Los niveles de coerción que usa R son los siguientes:

$$Lógico\rightarrow Entero\rightarrow Numérico\rightarrow Texto.$$

Es decir, si R debe transformar los elementos de un vector, los transformará al nivel más bajo de la escala anteriormente mostrada.

### Data frames

Los *data frames* son las estructuras más usadas en R para el análisis estadístico. Si el lector está familiarizado con el trabajo en Excel, los data frames son análogos a las tablas en el programa de Microsoft.

Básicamente, un data frame es un arreglo bidimensional (como una matriz) en el que las filas representan individuos y las columnas representan variables, las cuales pueden ser vistas como vectores. De esta forma, un data frame almacena diferentes tipos de datos relacionados a cada individuo.

En R, un data frame puede ingresarse de dos formas:

1. directamente, usando la función ```data.frame()``` e ingresando los datos manualmente, o
2. leyendo los datos de archivos externos, como una hoja de Excel u otro programa de hojas de cálculo similar.

También hay que agregar que R posee ejemplos de data frames ingresados por defecto que se pueden usar como ejemplos, algo que revisaremos más adelante.

**Ejercicio 1:** Ingresar un data frame que contenga la siguiente información:

| Nombre | Estatura (m) | Peso (kg) | Hijos | Edad (años) |
|---|---|---|---|---|
| Sofia | 1.65 | 60 | Sí | 32 |
| Alejandro | 1.80 | 75 | No | 28 |
| Isabella | 1.58 | 55 | Sí | 35 |
| Diego | 1.72 | 70 | No | 25 |
| Valentina | 1.60 | 58 | Sí | 30 |
| Javier | 1.75 | 72 | No | 29 |

**Solución.** Como mencionamos antes, un data frame puede se rvisto como una colección de vectores. Entonces ingresaremos cada variable como vector, y asignaremos el tipo de dato de acuerdo a su naturaleza.

Primero ingresaremos los nombres, que pueden ser vistos como datos tipo texto:

````r
nombres<-c("Sofía","Alejandro","Isabella","Diego","Valentina","Javier")
````

Luego la estatura, que es tipo numérico:

````r
estatura<-c(1.65,1.80,1.58,1.72,1.60,1.75)
#Atención: es importante notar que los decimales se separan con el punto (.) y los diferentes elementos de un vector con coma (,)
````

Ahora va el peso, que es tipo entero:

````r
peso<-c(60,75,55,70,58,72)
````

Continuamos con los hijos, en los cuales usaremos el tipo lógico:

````r
hijos<-c(TRUE,FALSE,TRUE,FALSE,TRUE,FALSE)
````

Y finalmente ingresamos la edad que es de tipo entero:

````r
edad<-c(32,28,35,25,30,29)
````

Finalmente, unificaremos todos estos datos en un data frame:

````r
datos1<-data.frame(Nombre=nombres,Estatura=estatura,Peso=peso,Hijos=hijos,Edad=edad)
datos1
````

Para terminar, hablaremos de algunas formas de acceder a la información dentro de los data frames. Se puede filtrar filas y columnas dentro de los data frame usando corchetes, considerándolo como una matriz. Así, la primera dimensión filtraría las filas y la segunda las columnas. Por ejemplo, si queremos solo la información de los primeros 3 individuos del data frame ingresado anteriormente, procedermos como sigue:

````r
datos1[1:3,]
````

Aquí, la notación ```1:3``` es una manera abreviada de ingresar los números del 1 al 3. En otro ejemplo, si queremos la información de todos los individuos, pero solo de las últimas 2 columnas, debemos hacer lo siguiente:



````r
datos1[,4:5]
````

Y finalmente, podemos combinar todo y presentar la información de nombre y edad de las mujeres de nuestro data frame, haciendo lo siguiente:

````r
datos1[c(1,3,5),c(1,5)]
````

Otra forma de acceder a las variables individuales es mediante el operador ```$```. Para ello, se debe usar el nombre que se asignó a la variable al declarar el data frame. Por ejemplo, si queremos los datos de la variable estatura:



````r
datos1$Estatura
````

Como se mencionó antes, cada variable de un data frame puede verse como un vector, y podemos usar los métodos propios de esta estructura.

Existen muchos más métodos para el manejo de data frames en R, pero para los objetivos de este curso basta con lo que se ha mencionado a este momento.

## Estadística descriptiva con R

Al realizar el estudio estadístico de variables, tradicionalmente se procede de acuerdo a dos paradigmas:

- calculando medidas estadísticas (tendencia central, variabilidad y posición),
- construyendo gráficos descriptivos de acuerdo a la naturaleza de los datos.

Repasaremos a continuación los métodos básicos para afrontar estas tareas en R, dado un data frame ya ingresado.

### Medidas estadísticas

Antes de iniciar con esta sección, hay qua aclarar dos puntos importantes:

- Si bien para el estudio de grandes cantidades de datos tradicionalmente se procede primero con la construcción de tablas de frecuencia, R calcula estas tablas de forma interna, así que no se necesita explicitarla para la construcción de gráficos o medidas.
- El cálculo de las medidas que se presentarán a continuación solo se puede realizar para variables numéricas.

#### Medidas de tendencia central

Las dos medidas de tendencia central más usadas son la *media aritmética* y la *mediana*. Dentro del ambiente de trabajo de R, las funciones para calcular estas dos medidas son, respectivamente, ```mean()``` y ```median()```, y se aplican sobre vectores. Por ejemplo, usando el data frame sobre el que venimos trabajando, calcularemos la media y mediana de la estatura.

````r
mean(datos1$Estatura)
median(datos1$Estatura)
````

#### Medidas de variabilidad

Las medidas de variabilidad más conocidas son el *rango*, la *varianza* y la *desviación estándar*. En R no existe una función definida para el rango, sin embargo se puede improvisar fácilmente al restar las funciones definidas para el máximo ```max()``` y el mínimo ```min()```. Para las otras dos medidas, las funciones definidas en R son ```var()``` y ```sd()``` para la varianza y la desviación estándar respectivamente. Hay que remarcar que tanto varianza como desviación estándar se calculan de forma muestral; es decir, se dividen para el número de datos menos 1.

Entonces si queremos calcular el rango, varianza y desviación estándar de la edad en el data frame ```datos1```, procedemos así:

````r
max(datos1$Edad)-min(datos1$Edad)
var(datos1$Edad)
sd(datos1$Edad)
````

#### Medidas de posición

Las medidas de posición (cuartiles, quintiles, deciles...) son casos particulares de los *percentiles*, que se pueden calcular fácilmente en R mediante la función ```quantile()```. Esta función recibe dos parámetros:

1. la variable a analizar,
2. el percentil a calcular, expresado como decimal.

Por defecto, y usando la estatura como ejemplo, si no proporcionamos percentiles, la función nos devolverá los cuartiles:


````r
quantile(datos1$Estatura)
````

Y si queremos un percentil preciso (por ejemplo, el 35%):

````r
quantile(datos1$Estatura,0.35)
````

También se le pueden dar varios valores como un vector. Por ejemplo, si queremos los percentiles 10, 40 y 85:

````r
quantile(datos1$Estatura,c(0.10,0.40,0.85))
````

### Gráficos descriptivos

Para describir de manera sencilla una sola variable, se pueden utilizar varios gráficos:

- el *gráfico de barras* o *histograma*, útil tanto para variables cualitativas y cuantitativas;
- el *gráfico de caja y bigotes*, exclusivamente usado para variables numéricas.

Si bien se suele usar otro gráfico para las variables cualitativas (el *gráfico de pastel*), no se recomienda su uso, prefiriéndose el uso del histograma en la medida de lo posible; no obstante, mostraremos la forma de construirlo en R.

Para poder mostrar más claramente el uso de estos gráficos, cargaremos un set de datos más amplio que está por defecto incluido en R.

````r
data(iris) # Cargamos el data set
head(iris) # Mostramos las primeras 6 filas del data frame
````

Este set de datos tiene la información de la longitud y ancho de los pétalos y los sépalos de flores de tres especies diferentes. Por tanto, tentemos 4 variables cuantitativa y una variable cualitativa.

#### Histograma

La función que permite la creación de histogramas en R es ```hist()```, como se muestra a continuación:

````r
hist(iris$Petal.Length)
````

Como se mencionó antes, también se puede construir gráficos de barras para variables cualitativas, usando la combinación de las funciones ```barplot()``` y ```table()```, que calcula las frecuencias de cada valor de la variable cualitativa.

````r
barplot(table(iris$Species))
````

#### Gráfico de caja y bigotes

Esté gráfico se usa para resumir la variabilidad de un variable numérica y en particular para detectar puntos atípicos. En R se realiza mediante la función ```boxplot()```:

````r
boxplot(iris$Petal.Length)
````

También puede recibir varias variables a la vez y colocar los gráficos de caja en una misma escala para compararlas:

````r
boxplot(iris[,1:4])
````

Con este gráfico, por tanto, ya podemos hablar de estadística descriptiva multivariante, un tema que se puede hablar en un curso mucho más amplio.

Finalmente, para la construcción de diagramas de pastel se usa la función ```pie()```, útil solo para variables cualitativas mediante el uso de ```table()```, como se ve a continuación:

````r
pie(table(iris$Species))
````
